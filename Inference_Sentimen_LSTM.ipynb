```python
import pickle
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load tokenizer dan model
with open("tokenizer_lstm.pkl", "rb") as f:
    tokenisasi_lstm = pickle.load(f)

model_lstm = load_model("model_lstm_wordembedding.h5")

# Parameter padding
panjang_maksimal_lstm = 50

# Mapping label
label_map = {0: 'negatif', 1: 'netral', 2: 'positif'}

# Contoh kalimat baru
kalimat_baru = [
    "Cukup membantu saya untuk bepergian di malam hari.",
    "Pelayanannya sangat ramah dan cepat, saya suka sekali.",
    "Aplikasinya sering error jadi males."
]

# Tokenisasi dan padding
urutan_baru = tokenisasi_lstm.texts_to_sequences(kalimat_baru)
urutan_baru_padded = pad_sequences(urutan_baru, maxlen=panjang_maksimal_lstm, padding="post", truncating="post")

# Prediksi
prediksi = model_lstm.predict(urutan_baru_padded)
kelas_prediksi = np.argmax(prediksi, axis=1)

# Tampilkan hasil
for kalimat, label in zip(kalimat_baru, kelas_prediksi):
    sentimen = label_map.get(label, 'unknown')
    print(f'"{kalimat}" -> {sentimen}')
```

```python
# Fungsi prediksi untuk 1 ulasan
def prediksi_sentimen(teks):
    urutan = tokenisasi_lstm.texts_to_sequences([teks])
    urutan_padded = pad_sequences(urutan, maxlen=panjang_maksimal_lstm, padding="post", truncating="post")
    hasil = model_lstm.predict(urutan_padded)
    label = np.argmax(hasil, axis=1)[0]
    return label_map.get(label, 'unknown')

# Contol lain input ulasan dari user
ulasan = input("Masukkan ulasan pelanggan: ")
hasil = prediksi_sentimen(ulasan)
print(f"Prediksi Sentimen: {hasil}")
```
