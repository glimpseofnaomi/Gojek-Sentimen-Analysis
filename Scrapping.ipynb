Pada proyek ini, saya akan melakukan analisis sentimen terhadap ulasan dan pendapat pengguna terkait aplikasi "Gojek" di Play Store. Analisis sentimen akan membantu kita untuk memahami bagaimana pengguna merasakan dan mengungkapkan pandangan mereka terhadap aplikasi ini.

```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews_all, Sort, reviews

# Mengambil semua ulasan dari aplikasi dengan ID 'com.gojek.app' di Google Play Store.
# Proses scraping mungkin memerlukan beberapa saat tergantung pada jumlah ulasan yang ada.
scrapreview = reviews_all(
    'com.gojek.app',          # ID aplikasi
    lang='id',             # Bahasa ulasan (default: 'en')
    country='id',          # Negara (default: 'us')
    sort=Sort.MOST_RELEVANT, # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=100000             # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
# Konversi ke DataFrame
df = pd.DataFrame(scrapreview)
df.head()
```

```python
# Simpan ke file CSV
df.to_csv("gojekdf.csv", index=False)
```
